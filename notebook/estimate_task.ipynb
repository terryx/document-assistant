```python
import numpy as np
sims = 100000

# define the range of tasks
tasks = [
    np.random.uniform(2, 6, sims),
    np.random.uniform(1, 3, sims),
    np.random.uniform(1, 3, sims),
    np.random.uniform(1, 3, sims),
    np.random.uniform(4, 13, sims),
]
```

```python
import matplotlib.pyplot as plt

duration = sum(tasks)
expected = 22

# Calculate histogram
counts, bins = np.histogram(duration, bins=30, density=True)
counts *= 100  # Convert to percentage

# Plot histogram
plt.figure(figsize=(6, 3))  # Adjusted figure size for better readability
plt.hist(bins[:-1], bins, weights=counts, alpha=0.6, color='blue', edgecolor='black')
plt.axvline(expected, color='r')

# Calculate the percentage of simulations where duration is greater than expected
percentage_above_expected = (duration > expected).sum() / sims * 100
plt.text(9, 11, f'Percentage > {expected}: {percentage_above_expected:.2f}%', color='black')

# Set labels
plt.xlabel('Duration')
plt.ylabel('Percentage')
plt.title('Histogram of Duration with Expected Value')

plt.show()

# Calculate the percentage of simulations where duration is greater than expected
percentage_above_expected = (duration > expected).sum() / sims * 100
print(f"Percentage of simulations where duration is greater than expected: {percentage_above_expected:.2f}%")
```
